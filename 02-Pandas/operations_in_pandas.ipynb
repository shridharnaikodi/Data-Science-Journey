```python
#Create a function that can accept salary and return the bonus of the salary
# salary less than or equal to 1500 -----> 10% bonus
# salary between 1501 and 6000 ----------> 5% bonus
# salary between 6001 and 9900 ----------> 2.5% bonus
# salary greater than 9900 --------------> 0 (no bonus)

def bonus(sal):
  if sal <=1500:
    return round(sal * 0.10,3)
  elif sal >=1501 and sal<=6000:
    return round(sal * 0.05,3)
  elif sal>=6001 and sal <=9900:
    return round(sal * 0.025,3)
  elif sal>9900:
    return 0
  else:
    return 0
bonus(9879)
```

```python
#Pandas also allows you to broadcast a function logic
#
# Goal : Create a new column named bonussal and calc bonus of each emp
#        based on their salary


# To broadcast a function in Pandas , you can use apply()
opdata['bounsal']=opdata['esal'].apply(bonus)
opdata
```
