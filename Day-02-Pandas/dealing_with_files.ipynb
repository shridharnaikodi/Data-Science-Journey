```python
import pandas as pd
```

```python
# Create a DataFrame
# 1. Using Collection Object
# 2. Loading a file
```

```python
#Using Collection Object
#emp (eid,ename,esal)


list1=[
    [1,"shri",900],
    [2,"jeeva",300],
    [3,"vikas",23332]


]
data = pd.DataFrame(list1)
print(data)
```

```python
# Dataframe contains two types of index
# 1. Row Index
# 2. Column Index

# Index starts from 0

data=pd.DataFrame(list1,columns=["no","name","sal"])
print(data)
```

```python
data.info()   # info()---can help us to find the missing values in the dataset and also the data types of the columns in the dataset.
```

```python
import numpy as np
list2=[
    [1,"shri",900],
    [2,"jeeva",np.nan],
    [3,"vikas",23332]
]
data=pd.DataFrame(list2,columns=["no","name","sal"])
print(data)
```
